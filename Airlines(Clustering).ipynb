## Kmeans Clustering

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
# load dataset
df = pd.read_csv(r"C:\DS Assignments\Assignment 7(Clustering)\Airlines.csv")
df
```

```python
df.info()
```

```python
df.shape
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_airlines_df = scaler.fit_transform(df.iloc[:,1:])
```

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:

wcss = []
for i in range(1, 13):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_airlines_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 13), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Basically Elbow plot/scree plot used to identify the best K-value, In this case the best 'k' values are k=5, k=7
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(scaled_airlines_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
df['clusterid_new'] = clusters_new.labels_
```

```python
#these are standardized values.
clusters_new.cluster_centers_
```

```python
airlines=df.drop(columns=['ID#'])
airlines.head(5)
```

```python
airlines.shape
```

```python
array=airlines.values
```

```python
#normalize data
from sklearn.preprocessing import StandardScaler
stsclr=StandardScaler().fit(airlines)
norm_data=stsclr.transform(airlines)
```

```python
#DBSCAN clustering
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=1.4,min_samples=12).fit(norm_data)
```

```python
#Noisy samples are given the label -1.
model.labels_
```

```python
clusters=pd.DataFrame(model.labels_,columns=['clusters'])
```

```python
clusters.value_counts()
```

```python
#-1 cluster represents outliers which means we have 440 datapts as outliers
final_data=pd.concat([clusters,airlines],axis=1)
final_data
```

```python
final_data.groupby(final_data.clusters).sum()
```
